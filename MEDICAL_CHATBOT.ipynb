```python
# FIXED Medical Chatbot - Complete Implementation for Google Colab
# This version fixes the prediction issues and ensures proper responses

# ============================================================================
# SECTION 1: INSTALLATION AND IMPORTS
# ============================================================================

# Install required packages
!pip install nltk tensorflow scikit-learn pandas numpy matplotlib seaborn
!pip install gradio

# Import libraries
import json
import pickle
import random
import numpy as np
import pandas as pd
import nltk
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import gradio as gr
import warnings
warnings.filterwarnings('ignore')

# Download NLTK data
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('omw-1.4')  # Additional download for better lemmatization
nltk.download('punkt_tab') # Download the missing resource

from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

print("All packages installed and imported successfully!")

# ============================================================================
# SECTION 2: ENHANCED MEDICAL TRAINING DATA
# ============================================================================

# Create more comprehensive training data with more patterns per intent
medical_intents = {
    "intents": [
        {
            "tag": "greeting",
            "patterns": [
                "Hi", "Hello", "Hey", "Good morning", "Good afternoon", "Greetings",
                "Hi there", "Hello there", "Hey there", "Good day", "Hi doctor",
                "Hello doctor", "Good evening", "How are you", "Hi bot", "Hello bot"
            ],
            "responses": [
                "Hello! I'm your medical assistant. How can I help you today?",
                "Hi there! I'm here to help with your medical questions.",
                "Greetings! How can I assist you with your health concerns?",
                "Hello! What health questions can I help you with today?"
            ]
        },
        {
            "tag": "goodbye",
            "patterns": [
                "Bye", "Goodbye", "See you later", "Thank you", "Thanks",
                "Bye bye", "See you", "Until next time", "Take care", "Farewell",
                "Thanks for help", "Thank you for helping", "That's all", "Nothing else"
            ],
            "responses": [
                "Take care of your health!",
                "Goodbye! Remember to consult a doctor for serious concerns.",
                "Stay healthy! Feel free to ask if you have more questions.",
                "Take care! Don't hesitate to reach out if you need help."
            ]
        },
        {
            "tag": "headache",
            "patterns": [
                "I have a headache", "My head hurts", "Headache relief", "Head pain", "Migraine",
                "Severe headache", "Bad headache", "Headache problems", "Head is pounding",
                "Sharp head pain", "Throbbing head", "Head pressure", "Tension headache",
                "Cluster headache", "Sinus headache", "My head is killing me"
            ],
            "responses": [
                "For headaches, try: 1) Rest in a quiet, dark room 2) Apply cold compress 3) Stay hydrated 4) Consider over-the-counter pain relievers like ibuprofen or acetaminophen. If severe or persistent, consult a doctor.",
                "Headache relief: Get adequate sleep, manage stress, stay hydrated. For frequent headaches, please see a healthcare provider.",
                "Try relaxation techniques, ensure proper posture, and avoid triggers like bright lights. Consult a doctor if headaches are severe or frequent."
            ]
        },
        {
            "tag": "fever",
            "patterns": [
                "I have fever", "High temperature", "Fever treatment", "I'm burning up", "Temperature",
                "Body temperature high", "Feeling feverish", "Hot body", "Chills and fever",
                "Fever and chills", "High fever", "Low grade fever", "Fever symptoms",
                "Body heat", "Warm forehead", "Thermometer shows fever"
            ],
            "responses": [
                "For fever: 1) Rest and stay hydrated 2) Use fever-reducing medication like acetaminophen or ibuprofen as directed 3) Apply cool compresses 4) Wear light clothing. Seek medical attention if fever exceeds 103Â°F (39.4Â°C) or lasts more than 3 days.",
                "Fever management: Drink plenty of fluids, rest, and monitor temperature. Contact a healthcare provider if fever is very high or accompanied by severe symptoms.",
                "Stay hydrated, rest, and use appropriate fever reducers. Seek immediate medical care for very high fever, difficulty breathing, or severe symptoms."
            ]
        },
        {
            "tag": "cold_flu",
            "patterns": [
                "I have a cold", "Flu symptoms", "Runny nose", "Cough", "Sore throat", "Stuffy nose",
                "Congestion", "Sneezing", "Blocked nose", "Throat pain", "Dry cough", "Wet cough",
                "Phlegm", "Mucus", "Nasal congestion", "Scratchy throat", "Hoarse voice",
                "Lost voice", "Chest congestion", "Common cold", "Seasonal flu", "Influenza"
            ],
            "responses": [
                "Cold/flu care: 1) Get plenty of rest 2) Stay hydrated with warm liquids 3) Use humidifier 4) Gargle with warm salt water for sore throat 5) Use over-the-counter medications for symptom relief. See a doctor if symptoms worsen or persist beyond 10 days.",
                "For cold/flu: Rest, fluids, and time are the best medicine. Use throat lozenges, warm liquids, and consider decongestants. Consult a doctor if symptoms persist or worsen.",
                "Common cold care: Rest, hydration, warm liquids, and symptom management. Most colds resolve in 7-10 days. See a healthcare provider if symptoms are severe or prolonged."
            ]
        },
        {
            "tag": "stomach_pain",
            "patterns": [
                "Stomach ache", "Abdominal pain", "Stomach hurts", "Belly pain", "Nausea",
                "Stomach cramps", "Digestive issues", "Indigestion", "Heartburn", "Acid reflux",
                "Stomach upset", "Gastric problem", "Abdominal cramps", "Stomach burning",
                "Vomiting", "Feel sick", "Queasy", "Motion sickness", "Food poisoning symptoms"
            ],
            "responses": [
                "For stomach pain: 1) Try bland foods (BRAT diet: bananas, rice, applesauce, toast) 2) Stay hydrated with clear fluids 3) Apply heat pad to abdomen 4) Avoid spicy, fatty, or dairy foods. Seek medical attention for severe pain, persistent vomiting, or signs of dehydration.",
                "Stomach discomfort: Rest, drink clear liquids like broth or herbal tea, eat light foods. Avoid dairy and spicy foods. Contact a doctor if pain is severe or accompanied by fever.",
                "For mild stomach issues: Drink clear fluids, eat light foods, and rest. Seek immediate medical care for severe pain, blood in stool/vomit, or persistent vomiting."
            ]
        },
        {
            "tag": "blood_pressure",
            "patterns": [
                "High blood pressure", "Hypertension", "Blood pressure", "BP", "Heart pressure",
                "Blood pressure high", "Elevated BP", "Systolic pressure", "Diastolic pressure",
                "Pressure readings", "BP medication", "Hypertensive", "Low blood pressure",
                "Hypotension", "Blood pressure control", "Monitoring BP"
            ],
            "responses": [
                "For high blood pressure: 1) Limit salt intake to less than 2300mg daily 2) Exercise regularly (30 minutes most days) 3) Maintain healthy weight 4) Limit alcohol 5) Don't smoke 6) Manage stress 7) Take prescribed medications consistently. Regular monitoring is essential.",
                "Blood pressure management: Follow a heart-healthy diet (DASH diet), exercise regularly, limit sodium, maintain healthy weight, and take medications as prescribed. Regular check-ups are important.",
                "Hypertension care: Lifestyle changes including low-sodium diet, regular exercise, stress management, and strict medication compliance. Work closely with your healthcare provider for regular monitoring."
            ]
        },
        {
            "tag": "diabetes",
            "patterns": [
                "Diabetes", "Blood sugar", "High glucose", "Diabetic", "Sugar levels",
                "Diabetes management", "Insulin", "Blood glucose", "Hyperglycemia", "Hypoglycemia",
                "Diabetic diet", "Sugar control", "Glucose monitoring", "Type 1 diabetes",
                "Type 2 diabetes", "Prediabetes", "Diabetic complications", "HbA1c"
            ],
            "responses": [
                "Diabetes management: 1) Monitor blood sugar regularly as directed 2) Follow prescribed meal plan 3) Take medications/insulin as directed 4) Exercise regularly (with doctor approval) 5) Practice proper foot care 6) Attend regular check-ups with your healthcare team.",
                "For diabetes: Maintain consistent meal times, monitor glucose levels, take medications properly, stay physically active, and keep regular medical appointments. Always carry glucose tablets for low blood sugar episodes.",
                "Diabetes care involves careful diet control, medication adherence, regular blood sugar monitoring, exercise, and working closely with your healthcare team for optimal management and preventing complications."
            ]
        },
        {
            "tag": "emergency",
            "patterns": [
                "Emergency", "Urgent", "Serious symptoms", "Cannot breathe", "Chest pain", "Severe pain",
                "Heart attack", "Stroke", "Difficulty breathing", "Unconscious", "Severe bleeding",
                "Overdose", "Suicide thoughts", "Can't breathe", "Choking", "Severe allergic reaction",
                "Anaphylaxis", "Call 911", "Need help now", "Medical emergency"
            ],
            "responses": [
                "ðŸš¨ MEDICAL EMERGENCY: If you're experiencing chest pain, difficulty breathing, severe bleeding, loss of consciousness, signs of stroke, or other serious symptoms, call emergency services (911 in US) immediately!",
                "âš ï¸ This sounds like an emergency. Please call 911 or go to the nearest emergency room immediately. Don't delay seeking professional medical help!",
                "ðŸ†˜ For medical emergencies, please call emergency services right away. This chatbot cannot replace emergency medical care! Call 911 now!"
            ]
        },
        {
            "tag": "mental_health",
            "patterns": [
                "Stress", "Anxiety", "Depression", "Mental health", "Feeling sad", "Worried",
                "Panic attack", "Anxious", "Overwhelmed", "Depressed", "Mood swings",
                "Emotional problems", "Stress management", "Mental wellness", "Feeling down",
                "Panic disorder", "Social anxiety", "Burnout", "Emotional support"
            ],
            "responses": [
                "Mental health is important. For stress/anxiety: 1) Practice deep breathing and relaxation techniques 2) Regular exercise 3) Adequate sleep (7-9 hours) 4) Talk to someone you trust 5) Consider professional counseling. Don't hesitate to seek help from mental health professionals.",
                "Mental wellness: Practice self-care, maintain social connections, exercise regularly, establish routines, and seek professional help when needed. Mental health is as important as physical health.",
                "For mental health concerns: Try mindfulness, deep breathing, regular routine, and professional support when needed. Remember, seeking help is a sign of strength, not weakness."
            ]
        },
        {
            "tag": "capabilities",
            "patterns": [
                "What can you help with", "What do you know", "What can you do", "Your capabilities",
                "What topics", "What conditions", "What can I ask", "Help me understand",
                "What are your limits", "What don't you know", "Limitations", "Scope"
            ],
            "responses": [
                "I can help with these medical topics:\nâœ… Headaches & migraines\nâœ… Fever & temperature issues\nâœ… Cold, flu, cough, sore throat\nâœ… Stomach pain, nausea, indigestion\nâœ… Blood pressure management\nâœ… Diabetes care & blood sugar\nâœ… Stress, anxiety, basic mental health\nâœ… Emergency recognition (directs to 911)\nâœ… General health & prevention tips\n\nâŒ I CANNOT help with: Skin conditions, eye/ear problems, bone/joint issues, women's/men's health, cancer, heart diseases, lung diseases, allergies, medication dosages, lab results, or any specialized medical conditions.",
                "My expertise covers common health concerns:\nâ€¢ Basic symptoms: headaches, fever, cold/flu\nâ€¢ Digestive issues: stomach pain, nausea\nâ€¢ Chronic conditions: diabetes, blood pressure\nâ€¢ Mental wellness: stress, anxiety management\nâ€¢ Emergency situations (I'll direct you to call 911)\nâ€¢ General health prevention\n\nI cannot provide advice on specialized conditions, medication interactions, specific diagnoses, or complex medical procedures. Always consult healthcare professionals for comprehensive care."
            ]
        },
        {
            "tag": "unsupported",
            "patterns": [
                "Rash", "Skin problem", "Eye issue", "Vision problem", "Hearing problem", "Ear infection",
                "Broken bone", "Fracture", "Arthritis", "Joint pain", "Back pain", "Pregnancy",
                "Menstrual", "Cancer", "Tumor", "Asthma", "Allergy", "STD", "Kidney problem",
                "Liver problem", "Heart disease", "Medication dosage", "Drug interaction", "Lab results"
            ],
            "responses": [
                "I'm sorry, but I cannot provide advice on this specific condition. My expertise is limited to: headaches, fever, cold/flu symptoms, stomach issues, diabetes, blood pressure, and basic mental health concerns. For your condition, please consult with a qualified healthcare professional who can provide proper diagnosis and treatment.",
                "This topic is outside my scope of knowledge. I'm designed to help with common conditions like headaches, fever, cold symptoms, digestive issues, diabetes management, blood pressure, and basic stress/anxiety. For specialized medical conditions, skin problems, injuries, or complex health issues, please see a doctor or specialist.",
                "I cannot assist with this medical concern as it's beyond my training. I can help with: fever, headaches, common cold symptoms, stomach problems, basic diabetes/blood pressure management, and general wellness. For other conditions, please consult healthcare professionals for accurate advice and treatment."
            ]
        }
    ]
}

# Save the training data
with open('medical_intents.json', 'w') as f:
    json.dump(medical_intents, f, indent=2)

print("Enhanced medical training data created successfully!")
print(f"Total intents: {len(medical_intents['intents'])}")
total_patterns = sum(len(intent['patterns']) for intent in medical_intents['intents'])
print(f"Total training patterns: {total_patterns}")

# ============================================================================
# SECTION 3: IMPROVED DATA PREPROCESSING
# ============================================================================

class MedicalChatbotPreprocessor:
    def __init__(self):
        self.lemmatizer = WordNetLemmatizer()
        self.stop_words = set(stopwords.words('english'))
        # Remove some medical-relevant stop words that might be important
        self.stop_words.discard('not')
        self.stop_words.discard('no')
        self.stop_words.discard('very')

        self.words = []
        self.classes = []
        self.documents = []
        self.ignore_words = ['?', '.', ',', '!', "'", '"', '(', ')', '[', ']']

    def preprocess_text(self, text):
        # Convert to lowercase and tokenize
        tokens = word_tokenize(text.lower())
        # Remove punctuation and lemmatize (keep some stop words for medical context)
        tokens = [self.lemmatizer.lemmatize(word) for word in tokens
                 if word.isalnum() and word not in self.ignore_words]
        return tokens

    def load_and_process_data(self, intents_file='medical_intents.json'):
        with open(intents_file, 'r') as f:
            intents = json.load(f)

        for intent in intents['intents']:
            for pattern in intent['patterns']:
                # Tokenize each pattern
                word_list = self.preprocess_text(pattern)
                self.words.extend(word_list)
                # Add documents with pattern and tag
                self.documents.append((word_list, intent['tag']))
                # Add classes
                if intent['tag'] not in self.classes:
                    self.classes.append(intent['tag'])

        # Remove duplicates and sort
        self.words = sorted(list(set(self.words)))
        self.classes = sorted(list(set(self.classes)))

        print(f"Documents: {len(self.documents)}")
        print(f"Classes: {len(self.classes)} - {self.classes}")
        print(f"Unique words: {len(self.words)}")

        return intents

    def create_training_data(self):
        training = []
        output_empty = [0] * len(self.classes)

        for doc in self.documents:
            bag = []
            pattern_words = doc[0]

            # Create bag of words
            for word in self.words:
                bag.append(1 if word in pattern_words else 0)

            # Create output row
            output_row = list(output_empty)
            output_row[self.classes.index(doc[1])] = 1

            training.append([bag, output_row])

        # Shuffle training data
        random.shuffle(training)
        training = np.array(training, dtype=object)

        # Split into X and y
        train_x = np.array(list(training[:, 0]))
        train_y = np.array(list(training[:, 1]))

        print(f"Training data created: X={train_x.shape}, y={train_y.shape}")
        return train_x, train_y

# Initialize preprocessor and process data
preprocessor = MedicalChatbotPreprocessor()
intents_data = preprocessor.load_and_process_data()
train_x, train_y = preprocessor.create_training_data()

# ============================================================================
# SECTION 4: IMPROVED MODEL TRAINING
# ============================================================================

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)

def create_model(input_shape, num_classes):
    """Create an improved model architecture"""
    model = keras.Sequential([
        layers.Dense(256, input_shape=(input_shape,), activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.3),

        layers.Dense(128, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.3),

        layers.Dense(64, activation='relu'),
        layers.Dropout(0.2),

        layers.Dense(num_classes, activation='softmax')
    ])

    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

# Build and train the model
input_shape = len(train_x[0])
num_classes = len(train_y[0])

print(f"Building model with input_shape={input_shape}, num_classes={num_classes}")

model = create_model(input_shape, num_classes)
model.summary()

# Train the model with callbacks
print("\nTraining the model...")

callbacks = [
    keras.callbacks.EarlyStopping(
        monitor='val_accuracy',
        patience=15,
        restore_best_weights=True,
        verbose=1
    ),
    keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=8,
        min_lr=0.0001,
        verbose=1
    )
]

history = model.fit(
    train_x, train_y,
    epochs=200,
    batch_size=16,
    validation_split=0.2,
    verbose=1,
    callbacks=callbacks
)

# Plot training history
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Print final metrics
final_acc = max(history.history['val_accuracy'])
final_loss = min(history.history['val_loss'])
print(f"\nBest Validation Accuracy: {final_acc:.4f}")
print(f"Best Validation Loss: {final_loss:.4f}")

# Save the model and preprocessor
model.save('medical_chatbot_model.h5')
pickle.dump(preprocessor.words, open('words.pkl', 'wb'))
pickle.dump(preprocessor.classes, open('classes.pkl', 'wb'))
pickle.dump(preprocessor, open('preprocessor.pkl', 'wb'))

print("Model and preprocessor saved successfully!")

# ============================================================================
# SECTION 5: FIXED CHATBOT CLASS
# ============================================================================

class MedicalChatbot:
    def __init__(self, model_path='medical_chatbot_model.h5'):
        # Load the trained model and preprocessor
        try:
            self.model = keras.models.load_model(model_path)
            self.words = pickle.load(open('words.pkl', 'rb'))
            self.classes = pickle.load(open('classes.pkl', 'rb'))
            self.preprocessor = pickle.load(open('preprocessor.pkl', 'rb'))
            self.intents = intents_data

            # Lower confidence threshold for better responses
            self.threshold = 0.3

            print(f"Chatbot loaded successfully!")
            print(f"Vocabulary size: {len(self.words)}")
            print(f"Number of classes: {len(self.classes)}")
            print(f"Classes: {self.classes}")

        except Exception as e:
            print(f"Error loading chatbot: {e}")

    def clean_up_sentence(self, sentence):
        """Tokenize and lemmatize the input sentence"""
        return self.preprocessor.preprocess_text(sentence)

    def bag_of_words(self, sentence):
        """Create bag of words array"""
        sentence_words = self.clean_up_sentence(sentence)
        bag = [0] * len(self.words)

        for w in sentence_words:
            for i, word in enumerate(self.words):
                if word == w:
                    bag[i] = 1

        return np.array(bag)

    def predict_class(self, sentence):
        """Predict the intent class"""
        bow = self.bag_of_words(sentence)
        res = self.model.predict(np.array([bow]), verbose=0)[0]

        # Get the highest probability prediction
        max_index = np.argmax(res)
        max_prob = res[max_index]

        # Always return the most likely class, even if below threshold
        if max_prob > self.threshold:
            return [{
                "intent": self.classes[max_index],
                "probability": str(max_prob)
            }]
        else:
            # Still return the best guess but mark as uncertain
            return [{
                "intent": self.classes[max_index],
                "probability": str(max_prob),
                "uncertain": True
            }]

    def get_response(self, intents_list):
        """Get response based on predicted intent"""
        if not intents_list:
            return "I'm sorry, I couldn't understand that. Could you please rephrase your question?"

        intent_data = intents_list[0]
        tag = intent_data["intent"]
        confidence = float(intent_data["probability"])

        print(f"DEBUG: Predicted intent: {tag}, Confidence: {confidence:.3f}")  # Debug info

        # Find the matching intent and get response
        for intent in self.intents["intents"]:
            if intent["tag"] == tag:
                response = random.choice(intent["responses"])

                # Add uncertainty note for low confidence
                if confidence < 0.5 or intent_data.get("uncertain", False):
                    response += "\n\n(Note: I'm not entirely certain about this response. Please consult a healthcare professional for accurate advice.)"

                return response

        # Fallback response
        return "I can help with medical questions about symptoms, conditions, and general health advice. Could you please be more specific about your concern?"

    def chat(self, message):
        """Main chat function with scope awareness"""
        if not message or not message.strip():
            return "Please ask me a medical question! Type 'What can you help with?' to see my capabilities."

        print(f"DEBUG: Input message: '{message}'")  # Debug info

        # Check for emergency keywords first
        emergency_keywords = ['emergency', 'urgent', 'chest pain', 'cannot breathe', 'can\'t breathe',
                            'severe bleeding', 'unconscious', 'overdose', 'heart attack', 'stroke',
                            'difficulty breathing', 'choking', 'anaphylaxis', 'call 911', 'need help now']

        message_lower = message.lower()
        if any(keyword in message_lower for keyword in emergency_keywords):
            return "ðŸš¨ MEDICAL EMERGENCY: If this is a medical emergency, please call your local emergency number (911 in the US) immediately or go to the nearest emergency room. This chatbot cannot provide emergency medical care!"

        # Check for unsupported condition keywords
        unsupported_keywords = ['rash', 'skin', 'acne', 'mole', 'eye', 'vision', 'ear', 'hearing',
                              'broken', 'fracture', 'arthritis', 'joint', 'back pain', 'pregnancy',
                              'pregnant', 'menstrual', 'cancer', 'tumor', 'asthma', 'allergy',
                              'allergic', 'std', 'kidney', 'liver', 'heart disease', 'medication dosage',
                              'drug interaction', 'lab results', 'blood test', 'x-ray', 'mri', 'ct scan']

        if any(keyword in message_lower for keyword in unsupported_keywords):
            return ("I'm sorry, but this condition is outside my scope of knowledge. I'm designed to help with:\n\n"
                   "âœ… Headaches, Fever, Cold/Flu symptoms, Stomach issues, Blood pressure, Diabetes, "
                   "Basic mental health (stress/anxiety), and General wellness.\n\n"
                   "For your specific concern, please consult with a qualified healthcare professional "
                   "who can provide proper diagnosis and treatment.\n\n"
                   "âš ï¸ Disclaimer: Always seek professional medical advice for health concerns.")

        # Get prediction and response
        try:
            intents = self.predict_class(message)
            response = self.get_response(intents)

            # Add medical disclaimer
            disclaimer = "\n\nâš ï¸ Disclaimer: This is an AI assistant providing general health information only. Always consult with qualified healthcare professionals for medical advice, diagnosis, or treatment."

            return response + disclaimer

        except Exception as e:
            print(f"ERROR in chat function: {e}")  # Debug info
            return ("I'm sorry, I encountered an error. Please try rephrasing your question or ask about "
                   "topics within my scope: headaches, fever, cold/flu, stomach issues, blood pressure, "
                   "diabetes, or basic mental health. For other conditions, consult a healthcare professional.")

# Initialize the chatbot
print("Initializing Medical Chatbot...")
chatbot = MedicalChatbot()

# ============================================================================
# SECTION 6: COMPREHENSIVE TESTING
# ============================================================================

def test_chatbot_comprehensive():
    """Test the chatbot with various queries"""
    test_cases = [
        # Greetings
        "Hello",
        "Hi there",
        "Good morning",

        # Medical symptoms
        "I have a terrible headache",
        "My head is pounding",
        "I have high fever",
        "I'm running a temperature",
        "I have a bad cough",
        "My throat is sore",
        "I have stomach pain",
        "My belly hurts",
        "I feel nauseous",

        # Conditions
        "I have diabetes",
        "My blood pressure is high",
        "I'm feeling very anxious",
        "I feel depressed",

        # Emergency
        "I have chest pain",
        "I can't breathe properly",

        # Goodbyes
        "Thank you",
        "Goodbye",

        # Random/unclear
        "What should I do?",
        "Help me",
        "I don't feel well"
    ]

    print("=== COMPREHENSIVE CHATBOT TESTING ===\n")

    for i, message in enumerate(test_cases, 1):
        print(f"Test {i}: {message}")
        try:
            response = chatbot.chat(message)
            # Print only first part of response (before disclaimer)
            main_response = response.split('\n\nâš ï¸')[0]
            print(f"Bot: {main_response}")
        except Exception as e:
            print(f"ERROR: {e}")
        print("-" * 80)

# Run comprehensive tests
test_chatbot_comprehensive()

# ============================================================================
# SECTION 7: IMPROVED GRADIO INTERFACE
# ============================================================================

def create_improved_gradio_interface():
    """Create an enhanced Gradio interface"""

    def chat_interface(message, history):
        """Enhanced chat interface function"""
        if not message or not message.strip():
            return history, ""

        try:
            # Get bot response
            bot_response = chatbot.chat(message)

            # Add to history
            history.append([message, bot_response])

            return history, ""

        except Exception as e:
            error_response = f"I'm sorry, I encountered an error: {str(e)}. Please try again or consult a healthcare professional."
            history.append([message, error_response])
            return history, ""

    def clear_chat():
        """Clear chat history"""
        return [], ""

    # Create enhanced Gradio interface
    with gr.Blocks(
        title="Medical Assistant Chatbot",
        theme=gr.themes.Soft(),
        css="""
        .gradio-container {
            max-width: 800px !important;
            margin: auto !important;
        }
        .chat-message {
            font-size: 14px !important;
        }
        """
    ) as demo:

        gr.Markdown(
            """
            # ðŸ¥ Medical Assistant Chatbot

            ## âœ… **I CAN HELP WITH THESE CONDITIONS:**
            - **Headaches & Migraines** - Head pain, tension headaches, migraine relief
            - **Fever & Temperature** - High fever, chills, temperature management
            - **Cold & Flu** - Runny nose, cough, sore throat, congestion, sneezing
            - **Stomach Issues** - Stomach pain, nausea, indigestion, heartburn
            - **Blood Pressure** - Hypertension management, BP control tips
            - **Diabetes** - Blood sugar management, diabetic care advice
            - **Mental Health** - Stress, anxiety, depression, basic wellness
            - **Emergencies** - Recognition & immediate 911 direction
            - **General Health** - Prevention tips, wellness advice

            ## âŒ **I CANNOT HELP WITH:**
            - Skin conditions, rashes, acne, moles
            - Eye/vision problems, ear/hearing issues
            - Bone fractures, joint pain, arthritis
            - Pregnancy, women's health, men's health
            - Cancer, heart diseases, lung diseases
            - Allergies, asthma, kidney/liver problems
            - Medication dosages, drug interactions
            - Lab test interpretations, specific diagnoses

            **ðŸš¨ For emergencies, call 911 immediately!**
            **ðŸ’¡ Ask "What can you help with?" for more details**
            """
        )

        chatbot_interface = gr.Chatbot(
            height=500,
            label="ðŸ’¬ Chat with Medical Assistant",
            show_label=True,
            elem_classes=["chat-message"]
        )

        with gr.Row():
            msg = gr.Textbox(
                placeholder="Describe your symptoms or ask a health question...",
                label="Your Message",
                scale=4,
                max_lines=3
            )
            send_btn = gr.Button("Send ðŸ’¬", scale=1, variant="primary")
            clear_btn = gr.Button("Clear ðŸ—‘ï¸", scale=1)

        # Enhanced examples with clear capability demonstration
        gr.Examples(
            examples=[
                "Hello, what can you help me with?",
                "I have a severe headache, what should I do?",
                "I have fever and chills",
                "My throat is sore and I'm coughing",
                "My stomach hurts after eating",
                "How do I manage high blood pressure?",
                "I'm feeling very anxious and stressed",
                "What are some tips to stay healthy?",
                "I have diabetes, any daily care tips?",
                "What are your capabilities and limitations?"
            ],
            inputs=msg,
            label="ðŸ’¡ Try these example questions (within my scope):"
        )

        # Event handlers
        msg.submit(chat_interface, [msg, chatbot_interface], [chatbot_interface, msg])
        send_btn.click(chat_interface, [msg, chatbot_interface], [chatbot_interface, msg])
        clear_btn.click(clear_chat, outputs=[chatbot_interface, msg])

        gr.Markdown(
            """
            ---
            ## ðŸ” **CHATBOT SCOPE SUMMARY:**
            **âœ… CAN HELP:** Headaches, Fever, Cold/Flu, Stomach Issues, Blood Pressure, Diabetes, Mental Health Basics, Emergencies, General Wellness

            **âŒ CANNOT HELP:** Skin conditions, Eye/Ear problems, Injuries, Pregnancy, Cancer, Allergies, Medication advice, Lab results, Specialized conditions

            ---
            **âš ï¸ Important Medical Disclaimer:**
            This AI assistant provides general health information for educational purposes only.
            It should not be used as a substitute for professional medical advice, diagnosis, or treatment.
            Always consult with qualified healthcare professionals for medical concerns.

            **ðŸ†˜ Emergency:** If you're experiencing a medical emergency, call your local emergency services immediately (911 in the US).
            """
        )

    return demo

# Create and launch the improved interface
print("Creating enhanced Gradio interface...")
demo = create_improved_gradio_interface()

# Launch the interface
print("\n" + "="*60)
print("ðŸš€ LAUNCHING MEDICAL CHATBOT INTERFACE")
print("="*60)
print("The chatbot is now ready and should respond properly!")
print("Interface will open in a new tab/window...")

demo.launch(
    share=True,  # Creates a public link
    debug=True,
    show_error=True,
    server_name="0.0.0.0",
    server_port=7860
)

print("\n" + "="*60)
print("âœ… MEDICAL CHATBOT SUCCESSFULLY DEPLOYED!")
print("="*60)
print("ðŸŽ‰ Your chatbot should now respond correctly to medical queries!")
print("ðŸ”§ If you still see generic responses, try restarting the runtime and running again.")
```
